```python
tit=pd.read_csv("Titanic-Dataset.csv")
print(tit.head())
```

```python
sns.barplot(data=tit,x="Survived",y="Age")
plt.show()
```

```python
avg_fare=tit.groupby("Sex")["Fare"].mean()
plt.plot(avg_fare.index,avg_fare.values,marker="o")
plt.xlabel("Sex")
plt.ylabel("Average Fare")
plt.show()
```

```python
columns=["Age","Fare","Survived"]
tit[columns].corr()
sns.heatmap(data=tit[columns].corr())
```
