```python
import re

dna = "ATGCATGCATGCNNATGC"

# Find all ATG patterns
matches = re.findall("ATG", dna)
print("ATG found:", len(matches), "times")

# Find invalid characters
invalid = re.findall("[^ATGC]", dna)
print("Invalid characters:", invalid)
```
